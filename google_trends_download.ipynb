```python
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='it', tz=-60)
```

```python
### keywords (kw_list)
# up to 5 different keywords

### categories (cat)
# - "health" -> 45
#    - "health conditions" -> 419
#       - "cold&flu" -> 629

### timezone (tz)
# central european -> +60

### search type (gprop)
# default to web searches
# can be images, news, youtube, froogle (google shopping)

### time (timeframe)
# ...

### place (geo)
# ...
```

```python
kw_list = ["influenza"]
cat = 629 #0

pytrends.build_payload(kw_list, cat=cat, 
                       timeframe='today 5-y', geo='IT', gprop='')
```

```python
df = pytrends.interest_over_time()
df.head()
```

```python
from matplotlib import pyplot as plt
plt.plot(df['influenza'])
```
